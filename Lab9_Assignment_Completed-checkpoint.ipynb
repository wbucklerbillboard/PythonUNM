## 3. Compute NDVI  (40 points)

Now that we have extracted the bands required to compute NDVI, we can use rasterio to load them and do some calculations with them thanks to numpy.

```python
import numpy as np
# convert the red and nir to numpy arrays with 
# type float for NDVI calculation
red = red.astype(float)
nir = nir.astype(float)
```

```python
# Allow division by zero
np.seterr(divide='ignore', invalid='ignore')
```

```python
# the NDVI formula is NDVI = (Near Infrared Band - Red Band)/(Near Infrared Band + Red Band)
# calculate the ndvi using red and nir
ndvi = (nir-red)/(nir+red)
```

```python
# Plot the ndvi dataset
pyplot.imshow(ndvi)
```

```python
# write the ndvi to a raster file in your computer drive.
# Define spatial characteristics of output object 
# (basically they are analog to the input)
kwargs = multiband.meta
# Update kwargs (change in data type)
kwargs.update(dtype=rasterio.float64, count = 1)
```

```python
# Opening a dataset in writing mode('w') using rasterio.open(),  
# other parameters are set using **kwargs
# hint: replace the ? with proper parameters
new_dataset = rasterio.open(r'C:\Users\casca\Downloads\SimpleAccess\L9\NDVI.tif', 'w', **kwargs)
```

```python
# write the data into the first band of new dataset using .write()
new_dataset.write(ndvi,1)
```

```python
# Sync data to disk and finish using .close()
new_dataset.close()
```
