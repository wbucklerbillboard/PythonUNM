## Task 1

Overlay Countries and Cities

```python
%matplotlib inline
# Import geopandas module
import geopandas as gpd
# Read build-in dataset countries and cities using gpd.read_file()
# from the gpd.datasets.get_path('naturalearth_lowres') and gpd.datasets.get_path('naturalearth_cities')
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
capitals = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
```

```python
# take a look at the columns of world and capitals using .head()
world.head()
#capitals.head()
```

```python
# Select European countries from world
euro_countries = world[world['continent']=='Europe']
#euro_countries
```

```python
# because overlay fucntion only works for polygon
# here we change the capitals to polygon
capitals['geometry'] = capitals.buffer(1)
capitals.head()
```

```python
# Overlay the two layers (euro_countries and capitals) using intersection
# to find the capitals in Europe. gpd.overlay(df1,df2,how = ‘intersection’)
# the output geodataframe will have attributes from both inputs.
euro_capitals = gpd.overlay(euro_countries, capitals,how='intersection')
euro_capitals.head()
```

## Task 2

Let’s aggregate our travel time data by car travel times, i.e. the grid cells that have the same travel time to Railway Station will be merged together.

```python
# Import necessary modules, e.g geopandas
# %matplotlib inline
import geopandas as gpd
from geopandas import GeoDataFrame
import shapely.speedups
shapely.speedups.disable()
%matplotlib inline
# Read file TravelTimes_to_5975375_RailwayStation.shp 
# in data folder using gpd.read_file()
ttgrid = gpd.read_file(r"C:\Users\casca\Downloads\SimpleAccess\L8\Data\TravelTimes_to_5975375_RailwayStation.shp")
ttgrid.plot()
```

```python
# take a look at the columns of data using .head()
# this is a shapefile of grid polygons, the car_r_t in each
# polygon means the driving time from that grid to the railway station
ttgrid.head()
```

```python
# create a choropleth map based on 'car_r_t' column
# hint: GeoDataFrame.plot(column = ?)
ttgrid.plot(column='car_r_t')
```

```python
# Let’s aggregate our travel time data by car travel times using dissolve()
result_aggregated = ttgrid.dissolve(by="car_r_t")
result_aggregated.head()
```

```python
# compare the number of polygons before and after the aggregation
# hint: using len(DataFrame) function.
print len(ttgrid)
print len(result_aggregated)
```

```python
# plot the aggregated results using a colormap 'YlOrRd'
result_aggregated.plot(cmap='YlOrRd')
```

## Task 3

Geometric Operations using Geopandas.

```python
# %matplotlib inline
import geopandas as gpd
# Import geopandas module
%matplotlib inline
# Read build-in dataset countries and capital cities using gpd.read_file()

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
capitals = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
```

```python
# Rotate the world map by 180 degree using GeoSeries.rotate(angle, origin='center')
# hint: the origin need to be (0,0).
world_r = world.rotate(180,origin=(0,0))
# plot the rotated world map
world_r.plot(cmap='Set1')
```

```python
# find the centroid of each country using .centroid
centroids = world.centroid
# create a 5 degree buffer on each centroid, using .buffer().
centroids_buffer = centroids.buffer(2)
centroids_buffer.plot()
```

```python
# plot the centroids buffer over the world map
# hint: using different color and set the alpha (transpanrenct of buffers)

base=world.plot()
centroids_buffer.plot(ax=base,alpha=.5, color='red')
```
