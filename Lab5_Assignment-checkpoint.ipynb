```python
# Import necessary modules first
# e.g. fiona, import geopandas as gpd;
# from shapely.geometry import Point,LineString, Polygon
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon
import fiona
```

```python
##  Read the data with Pandas into a DataFrame use pandas.read_csv()
my_data = pd.read_csv(r"C:\Users\casca\Downloads\SimpleAccess\some_posts.csv")
#my_data
#pd.DataFrame.from_csv("NYPD_Complaint_Data_Historic.csv")
```

```python
##  Create an empty column called geometry in my_data where you will store shapely Point objects
my_data['geometry'] = None
#my_data
```

```python
## Iterate over the rows of the DataFrame and insert Point objects into column
## geometry (you need to use .loc indexer to update the row, see materials
for index, row in my_data.iterrows():
    point = Point(row[0], row[1])
    my_data.loc[index, 'geometry'] = point
my_data    
#...'geometry'] = my_data.loc[index,['lat','lon']]    
#error    
```

```python
## Convert that DataFrame into a GeoDataFrame
## hint: gpd.GeoDataFrame(DataFrame,geometry = 'geometry column name')
geo = gpd.GeoDataFrame(my_data, geometry = 'geometry')
```

```python
##  Update the CRS for coordinate system as WGS84 (i.e. epsg code: 4326)
from fiona.crs import from_epsg
geo.crs = from_epsg(4326)
```

## 5. Create buffers around shopping centers (20 points)

Let's calculate a 5 km buffer around the shooping center points (shopping_centers.shp).

```python
# Import necessary modules
# e.g. import pandas as pd, geopandas as gpd;
# from shapely.geometry import Point,LineString, Polygon
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon
import fiona
```

```python
# read the shopping_centers.shp in data folder using gpd.readfile()
shpctrs = gpd.read_file(r"C:\Users\casca\Downloads\SimpleAccess\ShoCen\shopping_centers.shp")
```

```python
# Create a new empty column called 'buffer' to your shopping-centers GeoDataFrame
shpctrs['buffer'] = None
shpctrs
```

```python
## Iterate over the rows of the DataFrame and update the buffer column with a 
## 5 km buffer Polygon (because the unit is feet, so 16404.2 is the buffer distance)
## You need to use .loc indexer to update the row.
## Use Shapely's buffer function to create the buffer 
## (see http://toblerity.org/shapely/manual.html#object.buffer for details how to use it)
## You only need to use the distance -parameter, don't care about the other parameters.
#.buffer(16404.2)

for i, row in shpctrs.iterrows():
    pointbuff = shpctrs.loc[i, 'geometry']
    shpctrs.loc[i,'buffer'] = pointbuff.buffer(16404.2)
shpctrs
#need to clear buffer row
```

```python
# Replace the values in geometry column with the values of buffer column, and drop the buffer column.
shpctrs['geometry'] = shpctrs['buffer'] 
shpctrs = shpctrs.drop(['buffer', 'address_y'], axis = 1)
```

```python
# Have a look of the data you created by using GeoDataFrame.plot() -funtion.
%matplotlib inline
shpctrs.plot()
```

```python
##  Save the data into a Shapefile called shopping_centers_5kmbuffer.shp
shpctrs.to_file(r"C:\Users\casca\Downloads\SimpleAccess\shopping_centers_5kmbuffer.shp")
```
