 ## Task 2

Create a raster from scratch using numpy, rasterio

```python
# import the numpy library
import numpy
```

```python
# create a 10row*10column random number array using numpy.random.rand(?,?)
arr_random = numpy.random.rand(10,10)
```

```python
# import from_origin from the raster.transform
# import rasterio
from rasterio.transform import from_origin
import rasterio
```

```python
# create a affine transformation matrix can be computed 
# using a function in the rasterio.transform module
transform = from_origin(0.5, 9.5, 1, 1)
transform
```

```python
# Opening a dataset in writing mode using rasterio.open(), 
# with driver = 'GTiff', fill in the parameters
# height, width, count, dtype, crs='+proj=latlong', transform=transform
# height=Z.shape[0],width=arr_random.shape[1]
new_dataset = rasterio.open(r"C:\Users\casca\Downloads\SimpleAccess\L9\new.tif", 'w',driver="GTiff",height=arr_random.shape[0],width=arr_random.shape[1], count=1,dtype=arr_random.dtype, crs='+proj=latlong', transform=transform) 
```

```python
# write the data into the first band of new dataset using .write()
new_dataset.write(arr_random, 1)
```

```python
# Sync data to disk and finish using .close()
new_dataset.close()
```

```python
# Test the new raster file by open it and read the first band.  
# C:\Users\casca\Downloads\SimpleAccess\L9
rasterio.open(r'C:\Users\casca\Downloads\SimpleAccess\L9\new.tif').read(1)
```

## Task 3

Plot raster using rasterio and matplotlib

```python
# import the raster library
# import pyplot from matplotlib
# %matplotlib inline
import rasterio
from matplotlib import pyplot
%matplotlib inline
```

```python
# read the raster campustif.tif in data folder using rasterio.open()
dataset = rasterio.open(r"C:\Users\casca\Downloads\SimpleAccess\L9\campustif.tif")
```

```python
# plot band 2 as two dimentional data using matplotlib.pyplot.imshow(?).  
# Will accept only one band at a time.  Row and Column
# numbers are only displayed, not coordinates
pyplot.imshow(dataset.read(2))
```

Now we use the rasterio.plot.show(data) to display raster

```python
# import show from rastrio.plot
from rasterio.plot import show
```

```python
# show the 3 bands one by one using the rasterio.plot.show(?)
show(dataset.read(1))
show(dataset.read(2))
show(dataset.read(3))
```

The 3 bands are actually the red/green/blue channels. You can create a figure with  
multiple subplots by passing the show(..., ax=axis) argument.  
Also passing parameter to set a title for each subplot.

```python
# create a figure with 3 subplots.  Refer to lecture 4 or 5 for subplots.
fig, (axr, axg, axb) = pyplot.subplots(1,3, figsize=(21,7))
# show the band 1 in axr with colormap Reds
show(dataset.read(1), ax=axr, cmap='Reds', title='red channel')
# show the band 2 in axg with colormap Greens
show(dataset.read(2), ax=axg, cmap='Greens', title='green channel')
# show the band 3 in axb with colormap Blues
show(dataset.read(3), ax=axb, cmap='Blues', title='blue channel')
```
