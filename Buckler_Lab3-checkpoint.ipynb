### Step 4. Find 10 countries that have the most 'Quantity' ordered except UK

```python
# group by the Country (Hint. groupby() function,.groupby('Country').sum())
countrysum=online_rt.groupby('Country').sum()

# sort the value and get the first 10 after UK, using .sort_values()
firstten=countrysum.sort_values('Quantity', ascending=False)
firstten
# get the first 10 after UK (2nd - 11th in quantities)
dispfirst10=firstten[1:11]
dispfirst10
```
