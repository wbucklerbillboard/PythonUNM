```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}




### Step 3. Create 3 dataframes (data1, data2, data3) based on the 3 datasets provided
### e.g. data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data1=pd.DataFrame(raw_data_1, columns = ['subject_id','first_name','last_name'])
data2=pd.DataFrame(raw_data_2, columns = ['subject_id','first_name','last_name'])
data3=pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])
```

```python
### Step 4. Combine data1 and data2 using concat() to make a new dataframe called all_data
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

data1=pd.DataFrame(raw_data_1, columns = ['subject_id','first_name','last_name'])
data2=pd.DataFrame(raw_data_2, columns = ['subject_id','first_name','last_name'])
data3=pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])

data_1_2 = [data1, data2]
all_data = pd.concat(data_1_2)
all_data
```

```python
### Step 5. Join the two dataframes along columns and assing to all_data_col
all_data_col = pd.concat([data1, data2], axis = 1)
all_data_col
```

```python
### Step 6. Print data3
print(data3)
```

```python
### Step 7. Table join all_data and data3 using shared key subject_id (Hint: merge())

result = pd.merge(all_data, data3, on='subject_id')
result
```
